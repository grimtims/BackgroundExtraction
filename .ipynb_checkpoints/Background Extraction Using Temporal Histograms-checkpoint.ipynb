# Utility Functions

```python
def show_image(image):
    cv2.imshow('image',image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# high-performance counter
def most_common(lst):
    data = Counter(lst)
    return max(lst, key=data.get)
```

# Temporal Histogram Algorithm

### Slow, using python

```python
def dominant_pixel_value_slow(frameList):
    
    tempImage = np.zeros((len(frameList[0]), len(frameList[0][0])), dtype = np.uint8)
    
    for y in range(len(frameList[0])): # height
        
        for x in range(len(frameList[0][0])): # width
            
            # get the pixel value at position (x,y) of all frames in frameList
            posValues = [frameList[i][y][x] for i in range (len(frameList))]                
            
            # set image to most common pixel at position (x,y)
            tempImage[y][x] = int(most_common(posValues))
            
    return tempImage
```

### Fast, using C extensions

```python
%%cython --annotate
cimport cython
from libc.stdlib cimport malloc, free

#only uses 1 core though, could be parallelized, implementation is rudimentary 

# @cython.boundscheck(False)
cpdef unsigned char[:, :] dominant_pixel_value_fast(unsigned char [:, :, :] frameList):
     
    cdef int width, height, numPics, x, y, i
    cdef int domPixelWeight;
    cdef unsigned char domPixelValue;
    
    cdef unsigned char * posValues;
    cdef int hist_weights[256]
    cdef unsigned char output[360][640]
    cdef unsigned char test [360][640]
        
    numPics = frameList.shape[0]
    height = frameList.shape[1]
    width = frameList.shape[2]
    
    for y in range(0, height):
        
        for x in range(0, width):
            
            test[y][x] = frameList[numPics-1][y][x]
            
            ############# RESET COUNTERS ####################
            for l in range(0, 256):
                hist_weights[l] = 0;
            
            domPixelWeight = 0;
            domPixelValue = 0;
                      
            ###### ALLOCATE MEM FOR HISTOGRAM VALUES ######## 
            posValues = <unsigned char*>malloc(numPics*sizeof(unsigned char))
            for i in range(0, numPics):
                posValues[i] = frameList[i][y][x]
                        
            try:
            
            ########### COUNT WEIGHTS OF HISTOGRAM ###########
                for j in range(0, numPics):
                    hist_weights[posValues[j]] += 1
            
            ##### DETERMINE MAX WEIGHT & ASSIGN TO OUTPUT ####
                for k in range(0, 256):
                    if hist_weights[k] > domPixelWeight:
                        domPixelValue = k
                        domPixelWeight = hist_weights[k]

                output[y][x] = domPixelValue
                
            finally:
                free(posValues)
            
    return output
```

# Mean Filter

```python
%%cython --annotate
cimport cython
from libc.stdlib cimport malloc, free

#only uses 1 core though, could be parallelized, implementation is rudimentary 

# @cython.boundscheck(False)
cpdef unsigned char[:, :] mean_filter_fast(unsigned char [:, :] background, unsigned char [:, :]frame, unsigned int capIndex):
     
    cdef int width, height;
    cdef unsigned char output[360][640]
    cdef unsigned char test [360][640]
        
    height = frame.shape[0]
    width = frame.shape[1]
    
    for y in range(0, height):
        
        for x in range(0, width):
            
            if(capIndex != 0):
                output[y][x] = <unsigned char>(((background[y][x]*(capIndex-1)+frame[y][x]))/capIndex)

            
    return output
```

# Video Capture

```python
########## Video File ##############
videoFile = "german_highway.mp4"
# videoFile = "dehli_city.mp4"
# videoFile = "tokyo_night.mp4"
# videoFile = "ny_intersection.mp4"
# videoFile = "ny_timelapse.mp4"
# videoFile = "tokyo_inter.mp4"

cap = cv2.VideoCapture(videoFile);

###############################################################
#                     Video Properties                        #
###############################################################
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps    = cap.get(cv2.CAP_PROP_FPS)
totalFrames = cap.get(cv2.CAP_PROP_FRAME_COUNT)

print("Video Properties:\n")
print("\tFile: \t\t\t"+videoFile)
print("\tResolution: \t\t"+str(width)+" x "+str(height))
print("\tFPS: \t\t\t"+str(fps))
print("\tFrame Count: \t\t"+str(totalFrames))
###############################################################


#################### Initialize Displays ######################
cv2.namedWindow('Frames')
cv2.namedWindow('Background_Extraction')
###############################################################

captureInterval = 1/10 # in seconds
captureIndex = 0
print("\tCapture Interval: \t"+str(captureInterval)+" sec")

####### list of frame data
frameList = []

####### 2D aray for background image
backExtracImage = np.zeros((height, width), dtype = np.uint8)

bins = [x for x in range(0,256)]

while(cap.isOpened()):
    
    ##################################################
    #      Capture frame every 'frameInterval'
    ##################################################
    #time.sleep(captureInterval)
    captureIndex += 1;
    currentFrame = captureIndex*fps*captureInterval
    cap.set(cv2.CAP_PROP_POS_FRAMES, currentFrame)
    
    ret, frame = cap.read()
    if (ret == False or cv2.waitKey(1) & 0xFF == ord('q')):
        break

    # convert image to grayscale
    frameGray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    
    # add the frame to the list
    frameList.append(frameGray)
    
    ### Uncomment for python implementation ###
#     backExtracImage = dominant_pixel_value_slow(frameList)

    ### Uncomment for C implementation ###
    backExtracImage = np.asarray(dominant_pixel_value_fast(np.copy(frameList)))
    
    
    ####### MEAN FILTER
#     backExtracImage = np.asarray(mean_filter_fast(backExtracImage, frameGray, captureIndex))

            
    ##################################################
    #     Show Frames and Background Extraction 
    ##################################################
    
    totalFramesCap = int(((totalFrames/fps)/captureInterval)-1)
    temp_original = np.copy(frameGray) 
    cv2.putText(temp_original,"frame "+str(captureIndex)+"/"+str(totalFramesCap),(10,20),cv2.FONT_HERSHEY_SIMPLEX,0.800,(0,0,255),2)
    
    cv2.imshow('Frames',temp_original)
    cv2.imshow('Background_Extraction',backExtracImage)

    #print("frame "+str(captureIndex))
    ##################################################  
    
cv2.waitKey(0)
cap.release()
cv2.imwrite('extracted_background.png',backExtracImage)  
cv2.destroyAllWindows()
```

# Image Subtraction

```python
# Values from "Video Capture"

cap = cv2.VideoCapture(videoFile);

cv2.namedWindow('Original')
cv2.namedWindow('Background')
cv2.namedWindow('Difference')
cv2.namedWindow('Threshold')

def nothing(x):
    pass

cv2.createTrackbar('threshold','Threshold',0,255,nothing)

bins = [x for x in range(0,256)]

background = cv2.imread("extracted_background.png", 0)

while(cap.isOpened()):
    
    ret, frame = cap.read()
    if (ret == False or cv2.waitKey(1) & 0xFF == ord('q')):
        break
        
    time.sleep(0.1)

    # convert image to grayscale
    frameGray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    diff = cv2.absdiff(frameGray,background)
    
    
    threshImage = np.copy(diff)
    thresh = cv2.getTrackbarPos('threshold','Threshold')
    threshImage[threshImage > thresh] = 255;
    mask = cv2.erode(threshImage, None, iterations=2)
    mask = cv2.dilate(mask, None, iterations=2)
    
    cv2.imshow('Original',frameGray)
    cv2.imshow('Background', background)
    cv2.imshow('Difference', diff)
    cv2.imshow('Threshold', mask)


cv2.waitKey(0)
cap.release()
cv2.destroyAllWindows()
```
